# Checking and Handling Missing Values

```python
df.isnull()
```

```python
print("Missing Values:\n", df.isnull().sum())
```

```python
print("\nData Types:\n", df.dtypes)
```

```python
df.dropna(subset=["director", "cast"], inplace=True)
```

```python
df.fillna({"country":"Unknown"}, inplace=True)
```

```python

# df["rating"] = df["rating"].ffill()

df.fillna({"rating":df["rating"].mode()},inplace = True)
```

```python
df["date_added"] = df["date_added"].ffill()
```

```python
df.shape
```

# Finding and Removing Duplicates if any

```python
print(df.duplicated())
```

```python
print(df[df.duplicated()])
# if duplicates present :df.drop_duplicates(inplace=True)
```

# Standardizing Formats

```python
df['date_added'] = pd.to_datetime(df["date_added"], format='mixed')
```

```python
df["date_added"]
```

```python
df["release_year"] = df["release_year"].astype(int)
```

# Cleaning Text Data

```python
df["title"] = df["title"].str.strip().str.title()
df["director"] = df["director"].str.strip().str.title()
```

# Handling categorical inconsistencies:

```python
print(df["rating"].unique())
```

```python
ratings = {
    "TV-MA": "Mature",
    "R": "Restricted",
    "PG-13": "Teen",
    "TV-14": "Teen",
    "TV-G": "General",
    "TV-PG": "Parental Guidance",
    "NR": "Unrated",
    "PG": "Parental Guidance",
    "G": "General",
    "TV-Y7": "Children 7+",
    "TV-Y": "Children",
    "NC-17": "Adults Only",
    "TV-Y7-FV": "Children 7+",
    "UR": "Unrated"
}
```

```python
df["rating"] = df["rating"].replace(ratings)
```

```python
df.head()
```

```python
df.to_csv('CLEANED_Netflix_movies_and_tv_shows_clustering.csv', index=False)
print("Cleaned data saved as 'CLEANED_Netflix_movies_and_tv_shows_clustering.csv'")
```
